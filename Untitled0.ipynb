```python
import matplotlib.pyplot as plt
import numpy as np
mark1 = np.array([2, 4, 5, 7, 9, 10, 11, 13, 14, 15])
mark2 = np.array([6, 7, 8, 10, 11, 12, 13, 14, 15, 16])
plt.hist([mark1, mark2], bins=10, stacked=True, color=['skyblue','salmon'],edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Count')
plt.title('Stacked Histogram with Multiple Datasets')
plt.legend(['Mark Dataset 1','Mark Dataset 2'])
plt.show()
```

```python
import pandas as pd
a = [1, 7, 2]

myvar = pd.Series(a)
print(a)
print(myvar)
print('----------------------')
print(myvar[0])
print('----------------------')
myvar2 = pd.Series(a, index = ["x","y","z"])
print(myvar2)
print(myvar2["y"])
print('----------------------')
calories = {"day1": 420, "day2":380, "day3":390}
myvar = pd.Series(calories)
print(myvar)
print('----------------------')
myvar = pd.Series(calories, index = ["day1", "day2"])
print(myvar)
print('----------------------')
series= pd.Series(5, index=[0, 1, 2, 3])
print(series)
print('----------------------')
```

```python
int_series = pd.Series([1, 2, 3, 4])
print(int_series)
print('-------------------')
float_series = pd.Series([1.1, 2.2, 3.3, 4.4])
print(float_series)
print('-------------------')
string_series = pd.Series(['a', 'b', 'c', 'd'])
print(string_series)
print('-------------------')
data = {'a': 10, 'b': 20, 'c': 30}
series = pd.Series(data)
print(series['a'])
print(series[['a', 'c']])
print('-------------------')
data = [10, 20, 30, 40,50]
series = pd.Series(data)
print(series[series > 25])
print('-------------------')
```

```python
data = {
    'Name': ['Alice', 'Bod', 'Charlie', 'Alice', 'Bob'],
    'Age': [25, 30, 35, 25, 30],
    'city': ['New york', 'San Francisco', 'Los Andeles', 'chicago', 'UK'],
    'salary': [70000, 80000, 120000, 70000, 80000]
}

df = pd.DataFrame(data)
grouped = df.groupby('Name')
print(grouped.groups)
salary_sum = grouped['salary'].sum()
print(salary_sum)
age_mean = grouped['Age'].mean()
print(age_mean)
```
